```python
#Average Academic Performance by Addiction level

avg_academic = (df.groupby("Addiction_Level")["Academic_Performance"].mean().reset_index().sort_values("Addiction_Level").round(2))

print(avg_academic)
```

### Scatter Plot to display relationship between Addiction level to Academic Performance

***Bivariate Analysis***  
***X-Axis : Addiction level (Score:1-10)***  
***Y-Axis : Academic Performance (Out of 100)***


```python
#Scatter Plot to display relationship between addiction level to academic performance

plt.figure(figsize=(12,10))

plt.scatter(avg_academic['Addiction_Level'], avg_academic['Academic_Performance'], alpha =0.8)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.suptitle("Relationship between Addiction Level and Academic Performance", fontweight = 'bold', fontsize = 18, color = "#772222")
plt.title('(Survey based on 3000 Teenage Students over 3 months)', fontsize=11, color='black', pad=8)
plt.xlabel("Addiction Level (Score: 1-10)", fontsize = 14, fontweight = 'bold', color= "#772222")
plt.ylabel("Academic Performance (Out of 100)", fontsize = 14, fontweight = 'bold', color = "#772222")

plt.annotate('Outlier', xy=(2.6, 53), xytext=(3.5, 54), fontsize=11, color='gray', arrowprops=dict(arrowstyle='->', color='gray'))

plt.ylim(0,100)
plt.tight_layout()
plt.show()
```

Scatter plot shows no significant impact on Academic Performance due to addiction level. Most Academic Performance score ranged between 70-85. Therefore, no visible upward or downward trend. One outlier was identified(Score: 53)

```python
# Looking for column names to filter out the necessary numeric variables
df.columns
```

```python
# To find the correlation coefficient of the variables

df_corr = df[['Daily_Usage_Hours', 'Sleep_Hours', 'Academic_Performance', 'Parental_Control','Screen_Time_Before_Bed', 'Phone_Checks_Per_Day', 'Apps_Used_Daily',
       'Time_on_Social_Media', 'Time_on_Gaming', 'Time_on_Education','Family_Communication', 'Weekend_Usage_Hours',
       'Addiction_Level']].corr()   
df_corr  
```

### Heat Map to view the correlation of numeric variables

```python
# To see the correlation of numeric values

fig, ax = plt.subplots(figsize=(14, 14))

sns.heatmap(df_corr.round(2), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, annot_kws={'size': 10})
plt.suptitle('Teen Screentime Addiction Heatmap', fontsize=20, color= "#772222", fontweight = 'bold')
plt.title('(Survey based on 3000 Teenage Students over 3 months)', fontsize=11, color='black', pad=8)
plt.xticks(rotation=90, ha='center', fontsize=12)
plt.yticks(rotation=0, fontsize=12)

plt.tight_layout()
plt.show()
```

This Heatmap shows the correlation between the numeric variables. 

Red = positive correlation (both go up together)  
Blue = negative correlation (one goes up, other goes down)  
White/light = no relationship  
The diagonal is always 1.00 — every variable perfectly correlates with itself   
Daily_Usage_Hours has 0.60 is the Strongest finding means more screen time equal to higher addiction  
Next, Apps_Used_Daily has 0.32 means More apps is equal to slightly higher addiction  
Another important score is Sleep_Hours of -0.22 (not a strong range) indicates that more sleep may slightly reduce addiction 

### To Check Daily usage distribution - Histogram

```python
#To check the daily usage distribution

fig, ax = plt.subplots(figsize=(12, 12))  
ax.hist(df['Daily_Usage_Hours'], bins=20, alpha=0.9, color= 'steelblue', edgecolor = 'white')

fig.suptitle('Daily Usage Distribution', fontweight = 'bold', color= "#772222" , fontsize=20)
plt.title('(Survey based on 3,000 teenage students over 3 months)', fontsize=11, color='black', pad=8)
ax.set_xlabel('Daily phone usage (hours)', fontsize=15,fontweight = 'bold', color= "#772222" )  
ax.set_ylabel('Number of students', fontsize=15,fontweight = 'bold', color= "#772222" )         


mean_value = df['Daily_Usage_Hours'].mean()
ax.axvline(mean_value, color='red', linestyle='--', linewidth=1.5, label=f'Mean: {mean_value:.1f} hrs')
ax.legend()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

```

The histogram displays a univariate analysis of daily phone usage. The data appears balanced with no extreme skew. The most common usage is around 5 hours per day, with the majority of students falling within the 3–8 hour range.

```python
# To check any relationship between daily usage hour to addiction level
avg_addiction = df.groupby('Daily_Usage_Hours')['Addiction_Level'].mean().reset_index()
avg_addiction = avg_addiction[avg_addiction['Daily_Usage_Hours'] >= 0].sort_values('Daily_Usage_Hours').round(2)
print(avg_addiction)
```

### Average Addiction level Vs Daily Usage hours

```python
#Line Chart representing Average Addiction level to Daily Usage Hours

fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(avg_addiction['Daily_Usage_Hours'], avg_addiction['Addiction_Level'], linewidth=2, color='steelblue')

fig.suptitle('Daily Phone Usage (Hours) to Average Addiction Level', fontweight='bold', color="#772222", fontsize=18, y=0.98)
ax.set_title('(Survey based on 3,000 students over 3 months)', fontsize=12, color='gray')
ax.set_xlabel('Daily Phone Usage (Hours)', fontsize=15, color= "#772222", fontweight = 'bold')
ax.set_ylabel('Average addiction level (Out of 10)', fontsize=15, color= "#772222", fontweight = 'bold')

ax.yaxis.grid(True, linestyle='--', linewidth=0.4, color='gray', alpha=0.3)
ax.xaxis.grid(True, linestyle='--', linewidth=0.4, color='gray', alpha=0.3)
ax.set_axisbelow(True)


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlim(0,12)



plt.tight_layout()
plt.show()
```

This line chart confirms a strong positive relationship between daily phone usage hours and average addiction level. The overall trend clearly shows that as daily phone usage increases, addiction level rises consistently. There is steady clear upward trend from 2 to 8 hours. It also shows Addiction level starts around 4-5 hours and students using 8+ hours consistently scored the maximum addiction level.

```python
# To find out if lack of sleep impact academic performance

Sleep_Score = df.groupby('Sleep_Hours')['Academic_Performance'].mean().reset_index().round(2)
print(Sleep_Score)
```

### Correlation between Sleep Hours and Academic Performance

```python
# Line chart to find out if lack of sleep impacts academic performance

fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(Sleep_Score['Sleep_Hours'], Sleep_Score['Academic_Performance'], linewidth=2, color='steelblue')

fig.suptitle('Sleep Hours impact on Academic Performance', fontweight='bold', color="#772222", fontsize=18, y=.98)
ax.set_title('(Survey based on 3,000 students over 3 months)', fontsize=12, color='gray')
ax.set_xlabel('Sleep Hours', fontsize=15, color="#772222")
ax.set_ylabel('Academic Performance (Out of 100)', fontsize=15, color="#772222")

ax.yaxis.grid(True, linestyle='--', linewidth=0.4, color='gray', alpha=0.3)
ax.xaxis.grid(True, linestyle='--', linewidth=0.4, color='gray', alpha=0.3)
ax.set_axisbelow(True)


ax.set_ylim(0, 100)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

This line chart shows no clear trend between sleep hours and academic performance. Scores remained consistently within the 70–80 range regardless of sleep duration, confirming the weak correlation of -0.22 (weaker range) identified in the heatmap.
