```python
#groupby- Average Academic performance by Time_on_Education and Daily_Usage_Hours
result = (
    df.groupby(["Time_on_Education", "Daily_Usage_Hours"])["Academic_Performance"].mean().reset_index().sort_values("Academic_Performance"))

result
```

```python
#Average Anxiety level by Sleep hours and Addiction level
avg_anx = df.groupby(['Sleep_Hours','Addiction_Level'])['Anxiety_Level'].mean().reset_index().sort_index()
avg_anx
```
