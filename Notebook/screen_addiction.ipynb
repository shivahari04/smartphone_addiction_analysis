```python
#import libraries
 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data = pd.read_csv("../Data/teen_screentime_addiction.csv") # Reading CSV file
df = pd.DataFrame(data) 
```

```python
#groupby- Average Academic performance by Time_on_Education and Daily_Usage_Hours
result = (
    df.groupby(["Time_on_Education", "Daily_Usage_Hours"])["Academic_Performance"].mean().reset_index().sort_values("Daily_Usage_Hours"))

print(result)
```

```python
#Average Anxiety level by Sleep hours and Addiction level
avg_anx = df.groupby(['Sleep_Hours','Addiction_Level'])['Anxiety_Level'].mean().reset_index().sort_index().round(2)
avg_anx
```

```python
#Average Academic Performance by Addiction level

avg_academic = (df.groupby("Addiction_Level")["Academic_Performance"].mean().reset_index().sort_values("Addiction_Level").round(2))

avg_academic
```

### Scatter Plot to compare Addiction level  Vs Academic Performance

```python
plt.figure(figsize=(10,8))

plt.scatter(avg_academic['Addiction_Level'], avg_academic['Academic_Performance'])
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.title("Addiction Level Vs Academic Performance")
plt.xlabel("Addiction Level")
plt.ylabel("Academic Performance")

plt.tight_layout()
plt.show()
```

```python
df.columns
```

### Heat Map to compare all Numeric Values

```python
df_corr = df[['Daily_Usage_Hours', 'Sleep_Hours', 'Academic_Performance',
       'Social_Interactions', 'Exercise_Hours', 'Anxiety_Level',
       'Depression_Level', 'Self_Esteem', 'Parental_Control',
       'Screen_Time_Before_Bed', 'Phone_Checks_Per_Day', 'Apps_Used_Daily',
       'Time_on_Social_Media', 'Time_on_Gaming', 'Time_on_Education',
        'Family_Communication', 'Weekend_Usage_Hours',
       'Addiction_Level']].corr()   
df_corr  
```

```python
fig, ax = plt.subplots(figsize=(14, 14))

sns.heatmap(df_corr.round(2), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, ax=ax)
plt.title('Teen Screentime Addiction Heatmap', fontsize=22)
plt.xticks(rotation=90, ha='center')

plt.tight_layout()
plt.show()
```

### To Check Daily usage distribution - Histogram

```python
plt.Figure(figsize=(12,8))

plt.hist(df['Daily_Usage_Hours'],bins =20, alpha =0.9)

plt.title("Daily Usage Distribution", fontsize =16, pad=10)
plt.xlabel('Daily Phone usage Hours')
plt.ylabel('Frequency')

ax = plt.gca()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

### Average Addiction level Vs Daily Usage hours

```python
avg_addiction = df.groupby('Daily_Usage_Hours')['Addiction_Level'].mean().reset_index().round(2)
print(avg_addiction)
```

```python


plt.figure(figsize=(10, 5))
plt.plot(avg_addiction['Daily_Usage_Hours'], avg_addiction['Addiction_Level'])
plt.title('Average Addiction Level vs Daily Usage Hours')
plt.xlabel('Daily Usage Hours')
plt.ylabel('Average Addiction Level')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

### Correlation between Sleep Hours and Academic Performance

```python
Sleep_Score = df.groupby('Sleep_Hours')['Academic_Performance'].mean().reset_index().round(2)
print(Sleep_Score)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(Sleep_Score['Sleep_Hours'], Sleep_Score['Academic_Performance'])
plt.title('Academic Performance Vs Sleep Hours')
plt.xlabel('Sleep Hours')
plt.ylabel('Academic Perfomance')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
